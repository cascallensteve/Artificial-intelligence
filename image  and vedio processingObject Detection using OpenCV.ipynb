```python
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt
```

```python
image=cv2.imread("C:\\Users\\Comp science\\ai and ds files\\testimage.jpg")
```

```python
# Convert BGR to RGB for displaying using matplotlib
# blue green red to red green blue for displaying the image visualaiztion
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

step 3 Convert Image to Grayscale:

```python

# Convert to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Display the grayscale image
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Image')
plt.show()
```

4.	Apply Edge Detection:

```python

# Apply Canny edge detection
edges = cv2.Canny(gray_image, 50, 200)

# Display the edges
plt.imshow(edges, cmap='gray')
plt.title('Edge Image')
plt.show()
```

step5 Object Detection using Contours:

```python
# Find contours
contours, _ = cv2.findContours(edges, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

# Draw contours on the original image
image_contours = image.copy()
cv2.drawContours(image_contours, contours, -1, (0, 255, 0), 2)

# Convert BGR to RGB for displaying using matplotlib
image_contours_rgb = cv2.cvtColor(image_contours, cv2.COLOR_BGR2RGB)

```

```python
# Display the image with contours
plt.imshow(image_contours_rgb)
plt.title('Contours Image')
plt.show()
```
